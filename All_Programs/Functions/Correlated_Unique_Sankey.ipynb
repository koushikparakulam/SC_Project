```python
def Correlated_Unique_Sankey_Constructor(all_timepoints_cluster_labels, all_timepoint_correlated_unique_genes_dict, link_colors):

  unique_genes_all_link_value_based_colors = []
  unique_genes_source_nodes = []
  unique_genes_sink_nodes = []
  unique_genes_value_links = []


  all_timepoint_corr_uniq_genes_tuple_pairs = []

  for timepoints in all_timepoint_correlated_unique_genes_dict:

    timepoints_pairs_list = timepoints.split('_')
    cluster_pairs_dict = all_timepoint_correlated_unique_genes_dict.get(timepoints)

    unique_genes_all_link_value_based_colors, unique_genes_source_nodes, unique_genes_sink_nodes, unique_genes_value_links, all_timepoint_corr_uniq_genes_tuple_pairs = All_Source_Sink_Correlated_Unique_Constructor(cluster_pairs_dict, timepoints_pairs_list, all_timepoints_cluster_labels, unique_genes_source_nodes, unique_genes_sink_nodes, all_timepoint_corr_uniq_genes_tuple_pairs, unique_genes_all_link_value_based_colors, unique_genes_value_links, link_colors)

  return unique_genes_all_link_value_based_colors, unique_genes_source_nodes, unique_genes_sink_nodes, unique_genes_value_links, all_timepoint_corr_uniq_genes_tuple_pairs
```

```python
def Correlated_Unique_Value_Color_Links(total_correlated_genes, unique_genes_all_link_value_based_colors, unique_genes_value_links, link_colors):

  if total_correlated_genes >= 25:
      unique_genes_all_link_value_based_colors.append(link_colors[1])
      corr_total_adj = 9
      unique_genes_value_links.append(corr_total_adj)

  elif total_correlated_genes >= 15:
    unique_genes_all_link_value_based_colors.append(link_colors[2])
    corr_total_adj = 5
    unique_genes_value_links.append(corr_total_adj)

  elif total_correlated_genes >= 5:
    unique_genes_all_link_value_based_colors.append(link_colors[0])
    corr_total_adj = 3
    unique_genes_value_links.append(corr_total_adj)

  else:
    unique_genes_all_link_value_based_colors.append(link_colors[3])
    corr_total_adj = 1
    unique_genes_value_links.append(corr_total_adj)

  return unique_genes_all_link_value_based_colors, unique_genes_value_links
```

```python
def Modify_Label_Preset(cluster_pair, timepoints_pairs_list):

  cluster_pair_list = cluster_pair.split('__')

  cluster_1 = cluster_pair_list[0].replace('Cluster', '')
  cluster_1 = cluster_1.replace('_','')
  cluster_1_modified = cluster_1+'_'+timepoints_pairs_list[0]+'_'+'neuronal'

  cluster_2 = cluster_pair_list[1].replace('Cluster', '')
  cluster_2 = cluster_2.replace('_','')
  cluster_2_modified = cluster_2+'_'+timepoints_pairs_list[1]+'_'+'neuronal'

  return cluster_1_modified, cluster_2_modified
```

```python
def All_Source_Sink_Correlated_Unique_Constructor(cluster_pairs_dict, timepoints_pairs_list, all_timepoints_cluster_labels, unique_genes_source_nodes, unique_genes_sink_nodes, all_timepoint_corr_uniq_genes_tuple_pairs, unique_genes_all_link_value_based_colors, unique_genes_value_links, link_colors):

  for cluster_pair in cluster_pairs_dict:

    cluster_1, cluster_2 = Modify_Label_Preset(cluster_pair, timepoints_pairs_list)

    all_timepoint_corr_uniq_genes_tuple_pairs.append((cluster_1, cluster_2))

    correlated_genes_list = cluster_pairs_dict.get(cluster_pair)
    total_correlated_genes = len(correlated_genes_list)

    unique_genes_source_nodes.append(all_timepoints_cluster_labels.index(cluster_1))
    unique_genes_sink_nodes.append(all_timepoints_cluster_labels.index(cluster_2))

    unique_genes_all_link_value_based_colors, unique_genes_value_links = Correlated_Unique_Value_Color_Links(total_correlated_genes, unique_genes_all_link_value_based_colors, unique_genes_value_links, link_colors)

  return unique_genes_all_link_value_based_colors, unique_genes_source_nodes, unique_genes_sink_nodes, unique_genes_value_links, all_timepoint_corr_uniq_genes_tuple_pairs
```

```python
def Grey_Out_Non_Unique_Correlations(grey_out, source_nodes, all_link_value_based_colors, all_timepoints_cluster_labels, sink_nodes, all_timepoint_corr_uniq_genes_tuple_pairs):

  all_link_value_based_colors_non_unique_correlations = []

  for index in range(len(source_nodes)):

    linkage_tuple = (all_timepoints_cluster_labels[source_nodes[index]], all_timepoints_cluster_labels[sink_nodes[index]])

    if linkage_tuple in all_timepoint_corr_uniq_genes_tuple_pairs:
      all_link_value_based_colors_non_unique_correlations.append(all_link_value_based_colors[index])

    else:
      all_link_value_based_colors_non_unique_correlations.append(grey_out)

  return all_link_value_based_colors_non_unique_correlations
```
