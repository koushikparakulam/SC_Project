```python
def Harmony_Cluster_Labels_Constructor(timepoints_anndata, timepoints_anndata_names, E_I_M_labels_dict):

  for anndata, anndata_name in zip(timepoints_anndata, timepoints_anndata_names):
    anndata.obs['cluster_labels'] = ''
    leiden_E_I_M_conversion_dict = E_I_M_labels_dict.get(anndata_name)
    for transcripts in anndata.obs.index:
      full_E_I_M = leiden_E_I_M_conversion_dict.get(anndata.obs.loc[transcripts]['leiden'])
      anndata.obs.loc[transcripts, 'cluster_labels'] = full_E_I_M
```

```python
def E_I_M_Labels_Constructor(timepoints_anndata, timepoints_anndata_names, E_I_M_labels_dict):
  for anndata, anndata_name in zip(timepoints_anndata, timepoints_anndata_names):
    anndata.obs['E_I_M'] = ''
    leiden_E_I_M_conversion_dict = E_I_M_labels_dict.get(anndata_name)
    for transcripts in anndata.obs.index:
      full_E_I_M = leiden_E_I_M_conversion_dict.get(anndata.obs.loc[transcripts]['leiden'])
      partial_E_I_M = full_E_I_M.split('_')[0]
      anndata.obs.loc[transcripts, 'E_I_M'] = partial_E_I_M
```

```python
def Inhib_Excitatory_Anndata_Labels_Constructor(timepoints_anndata, Inhibitory, Excitatory, Mix):

  for anndata in timepoints_anndata:

    anndata.obs['Inhibitory_Excitatory_Mix'] = ''

    for transcripts in anndata.obs.index:
      if anndata.obs.loc[transcripts]['cluster_labels'] in Inhibitory:
        anndata.obs.loc[transcripts, 'Inhibitory_Excitatory_Mix'] = 'Inhibitory'

    for transcripts in anndata.obs.index:
      if anndata.obs.loc[transcripts]['cluster_labels'] in Excitatory:
        anndata.obs.loc[transcripts, 'Inhibitory_Excitatory_Mix'] = 'Excitatory'

    for transcripts in anndata.obs.index:
      if anndata.obs.loc[transcripts]['cluster_labels'] in Mix:
        anndata.obs.loc[transcripts, 'Inhibitory_Excitatory_Mix'] = 'Mix'
```

```python
def True_Relative_Non_Distinct_Labels_Constructor(timepoints_anndata, timepoints_anndata_names, all_timepoint_distinction_dict):

  for anndata, anndata_name in zip(timepoints_anndata, timepoints_anndata_names):

      anndata.obs['Distinction'] = ''

      true_distinct_clusters = all_timepoint_distinction_dict.get(anndata_name)[0]
      relative_distinct_clusters = all_timepoint_distinction_dict.get(anndata_name)[1]
      non_distinct_clusters = all_timepoint_distinction_dict.get(anndata_name)[2]


      for transcripts in anndata.obs.index:
        if true_distinct_clusters:
          if anndata.obs.loc[transcripts]['E_I_M'] in true_distinct_clusters:
            anndata.obs.loc[transcripts,'Distinction'] = 'True_Distinct'

        if relative_distinct_clusters:
          if anndata.obs.loc[transcripts]['E_I_M'] in relative_distinct_clusters:
            anndata.obs.loc[transcripts,'Distinction'] = 'Relative_Distinct'

        if non_distinct_clusters:
          if anndata.obs.loc[transcripts]['E_I_M'] in non_distinct_clusters:
            anndata.obs.loc[transcripts,'Distinction'] = 'Non_Distinct'
```
