```python
def All_Timepoint_Cluster_Labels_Constructor(timepoints_anndata, timepoints_anndata_names):

  all_timepoints_cluster_labels = []

  for anndata, name in zip(timepoints_anndata, timepoints_anndata_names):

    all_clusters = anndata.obs['cluster_labels'].unique().tolist()
    # original key=lambda x: int(x.split('_')[0].split('C')[1]
    all_clusters_sorted = sorted(all_clusters, key=lambda x: x.split('_')[0])

    curr_timepoint_labels = []

    for clust_label in all_clusters_sorted:

      curr_timepoint_labels.append(clust_label)
      all_timepoints_cluster_labels.append(clust_label)

  return all_timepoints_cluster_labels
```

```python
def General_Type_Color_Labels_Constructor(all_timepoints_cluster_labels, Inhib_Excit_Mix, Inhib_Excit_Mix_rgb):

  general_type_color_labels_index = []
  general_type_color_labels = []


  for general_type_rbg, general_type_tag in zip(Inhib_Excit_Mix_rgb, Inhib_Excit_Mix):

    for index, cell_tag in enumerate(all_timepoints_cluster_labels):

      if cell_tag in general_type_tag:
        general_type_color_labels_index.append((general_type_rbg, index))

  sorted_general_type_color_labels_index = sorted(general_type_color_labels_index, key=lambda t: t[1], reverse = False)
  general_type_color_labels = [x[0] for x in sorted_general_type_color_labels_index]

  return general_type_color_labels
```
