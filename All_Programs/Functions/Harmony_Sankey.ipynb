```python
def All_Sink_Weight_Tuples_Constructor(all_relations_list, all_timepoints_cluster_labels):

  unique_indicies = []
  all_sink_weight_tuples = []

  for relations in all_relations_list:

    weight_0 = relations[0][1]
    weight_1 = relations[1][1]
    weight_adj = min(weight_0,weight_1)
    max_weight = weight_adj

    sink_index = all_timepoints_cluster_labels.index(relations[1][0])
    sink_weight_tuple = (sink_index, weight_adj)

    min_weight_list = []
    label_indicies = []

    unique_indicies.append(sink_index)
    all_sink_weight_tuples.append(sink_weight_tuple)

    if sink_index in unique_indicies:
      for x, sink_tuples in enumerate(all_sink_weight_tuples):
        if sink_tuples[0] == sink_index:
          min_weight_list.append(sink_tuples[1])
          label_indicies.append(x)
      max_weight = max(min_weight_list)

      for y, ind in enumerate(label_indicies):
        if y == 0:
          all_sink_weight_tuples[ind] = (sink_index, max_weight)
        else:
          all_sink_weight_tuples.pop(ind)

  return all_sink_weight_tuples
```

```python
def Sink_Source_Value_Constructor(all_sink_weight_tuples, link_colors, all_timepoints_cluster_labels, relationship_tuple, harmony_all_link_value_based_colors, harmony_source_nodes, harmony_sink_nodes, harmony_value_links):

  for sink_weight in all_sink_weight_tuples:

    harmony_sink_nodes.append(sink_weight[0])
    harmony_value_links.append(sink_weight[1])

    if sink_weight[1] >= 70:
      harmony_all_link_value_based_colors.append(link_colors[1])

    elif (sink_weight[1] >= 40) and (sink_weight[1] < 70):
      harmony_all_link_value_based_colors.append(link_colors[2])

    elif (sink_weight[1] >= 20) and (sink_weight[1] < 40):
      harmony_all_link_value_based_colors.append(link_colors[0])

    else:
      harmony_all_link_value_based_colors.append(link_colors[3])

  harmony_source_nodes.extend([all_timepoints_cluster_labels.index(relationship_tuple) for x in range(len(all_sink_weight_tuples))])

  return harmony_all_link_value_based_colors, harmony_source_nodes, harmony_sink_nodes, harmony_value_links
```

```python
def Harmony_Sankey_Constructor(all_timepoints_relationship_tuples_dict, timepoint_pair_name_tuples, link_colors, all_timepoints_cluster_labels):

  harmony_all_link_value_based_colors = []
  harmony_source_nodes = []
  harmony_sink_nodes = []
  harmony_value_links = []

  for all_timepoints_integrated_label, timepoint_label_name in zip(all_timepoints_relationship_tuples_dict, timepoint_pair_name_tuples):

    relationship_tuples_dict = all_timepoints_relationship_tuples_dict.get(all_timepoints_integrated_label)

    for relationship_tuple in relationship_tuples_dict:

      if relationship_tuple.split('_')[1] == timepoint_label_name[0]:

        if relationship_tuple in relationship_tuples_dict:

          all_relations_list = relationship_tuples_dict.get(relationship_tuple)
          all_sink_weight_tuples = All_Sink_Weight_Tuples_Constructor(all_relations_list, all_timepoints_cluster_labels)

          harmony_all_link_value_based_colors, harmony_source_nodes, harmony_sink_nodes, harmony_value_links = Sink_Source_Value_Constructor(all_sink_weight_tuples, link_colors, all_timepoints_cluster_labels, relationship_tuple, harmony_all_link_value_based_colors, harmony_source_nodes, harmony_sink_nodes, harmony_value_links)

  return harmony_all_link_value_based_colors, harmony_source_nodes, harmony_sink_nodes, harmony_value_links
```
